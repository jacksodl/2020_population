# 2020 Population Analysis

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from matplotlib.pyplot import figure
# from matplotlib.ticker import FormatStrFormatter, StrMethodFormatter
import os
```

### Upload data

```python
population = pd.read_csv('./population_by_country_2020.csv')

# Rename columns to 'Population' and 'Country'
population = population.rename(columns={
    'Country (or dependency)' : 'Country',
    'Population (2020)' : 'Population'
})
# population.head(20)
```

### Add Average column for average population

```python
population['Average'] = population['Population'].mean()
population['Average'] = population['Average'].astype(int)
# Turn 'Average' column into an a whole number integer
population.head(40)
```

## Check and inspect data

```python
population.info()
# We have 235 rows and 12 columns
population.sample(n=50)
# Get random sample of 50 to see how data looks
```

```python
population.describe()
# Show statistical data of all the numeric columns
```

```python
population.value_counts().count()
# Unique values
```

```python
population.shape[0]
# Check number of rows
```

```python
population.shape[1]
# Check number of columns
```

## Top 15 most populous countries in the World

```python
top_15 = population.sort_values(by='Population', ascending=False).head(15)
top_15
```

```python
countries = population['Country'].head(15)
pop = population['Population'].head(15)
avg_pop = population['Average'].head(15)

fig, ax1 = plt.subplots(figsize=(15, 10))
# plt.hexbin(countries, , cmap=plt.cm.gnuplot2_r) # Commas in Y axis

ax2 = ax1.twinx()
ax1.bar(countries, pop, color='b')
ax2.plot(countries, avg_pop, 'r-', linewidth=5)

ax1.set_xlabel('Country', color='g', size = 15, weight='bold')
ax1.set_ylabel('Population', color='b', size = 15, weight='bold')
ax1.ticklabel_format(style='plain', axis='y')
ax2.set_ylabel('Average Population', color='r', size = 15)
ax1.set_xticklabels(countries, rotation = 55, weight='bold')
ax1.set_yticklabels(pop, weight='bold')
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda y,  p: format(int(y), ',')))
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda y,  p: format(int(y), ',')))
plt.title('2020 Population', size=30, weight='bold')

plt.show()


# plt.figure(figsize=(15,7))
# plt.bar(x, y)
# plt.xlabel('Country')
# plt.ylabel('Population')
# plt.xticks(x, rotation=40)
# plt.ticklabel_format(style='plain', axis='y')

# plt.show()
```
